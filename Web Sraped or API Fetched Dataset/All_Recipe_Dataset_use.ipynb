```python
# Clean 'Servings' column: convert to numeric (if needed), and drop non-convertibles
df['Servings'] = pd.to_numeric(df['Servings'], errors='coerce')
df = df.dropna(subset=['Servings'])

# Filter servings between 1 to 10
serving_range = df[(df['Servings'] >= 1) & (df['Servings'] <= 10)]

# Count frequency of each serving size
serving_counts = serving_range['Servings'].value_counts().sort_index()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(serving_counts, labels=serving_counts.index.astype(int), autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Servings (1 to 10)')
plt.axis('equal')  # Equal aspect ratio for a perfect circle
plt.show()
```

```python
import re

def time_to_minutes(time_str):
    if pd.isnull(time_str):
        return None
    time_str = time_str.lower()
    hours = re.search(r'(\d+)\s*hr', time_str)
    minutes = re.search(r'(\d+)\s*min', time_str)

    total_minutes = 0
    if hours:
        total_minutes += int(hours.group(1)) * 60
    if minutes:
        total_minutes += int(minutes.group(1))
    return total_minutes if total_minutes > 0 else None


df['Prep_Min'] = df['Prep Time'].apply(time_to_minutes)
df['Cook_Min'] = df['Cook Time'].apply(time_to_minutes)

df_valid = df.dropna(subset=['Prep_Min', 'Cook_Min'])

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df_valid['Prep_Min'], df_valid['Cook_Min'], alpha=0.7)
plt.xlabel('Prep Time (minutes)')
plt.ylabel('Cook Time (minutes)')
plt.title('Prep Time vs Cook Time')
plt.grid(True)
plt.show()
```
